# Define
Merge the 4 dog stages and some have more than 1 dog stage

```python
final['add_all'] = final.doggo + final.floofer + final.pupper + final.puppo
```

```python
#  to check dog stages
def check_stages(stages):
    if stages['add_all'].count('None') == 2:
        return 'Multiple' 
    else:
        if stages['add_all'].count('doggo') == 1:
            return 'Doggo'
        elif stages['add_all'].count('floofer') == 1:
            return 'Floofer'
        elif stages['add_all'].count('pupper') == 1:
            return 'Pupper'
        elif stages['add_all'].count('puppo') == 1:
            return 'Puppo'
        else:
            return 'None'

final['dog_stage'] = final.apply(check_stages, axis=1)
```

```python
final.info()
```

# Define
Noticing some URLs are longer than others

```python
#find length of short url
url_length = len(final['text'][0].split()[-1])+1
url_length

# add column with new short URL
final['url'] = final['text'].apply(lambda row: row[-url_length:])

#remove the short URL from text
final['text'] = final['text'].apply(lambda row: row[:-url_length])

```
